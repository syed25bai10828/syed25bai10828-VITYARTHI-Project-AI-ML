```python
import time

class IndustrialFirewall:
    def __init__(self):

        # Logic: From state 'OFF', you can only go to 'STARTING'.
        # From 'RUNNING', you can 'OPEN_VALVE' or 'STOP'.
        self.valid_transitions = {
            "OFF": ["STARTING"],
            "STARTING": ["RUNNING", "OFF"],
            "RUNNING": ["OPEN_VALVE", "STOPPING", "OFF"],
            "OPEN_VALVE": ["CLOSE_VALVE", "EMERGENCY_STOP"],
            "CLOSE_VALVE": ["RUNNING", "STOPPING"],
            "STOPPING": ["OFF"],
            "EMERGENCY_STOP": ["OFF"]
        }
        self.current_state = "OFF"
        self.log_history = []

    def inspect_packet(self, command):
        """
        Analyzes the incoming command against the logical State Machine.

        """
        print(f"\n[NETWORK] Incoming Command: {command}")

        if command in self.valid_transitions.get(self.current_state, []):
            # Update state if command is logical
            old_state = self.current_state
            self.current_state = command
            print(f"[SAFE] Transition: {old_state} -> {self.current_state}")
            return True
        else:
            # Detect Logic Anomaly
            print(f"[!!! ALARM !!!] Logic Violation Detected!")
            print(f"Reason: Cannot execute '{command}' while system is in '{self.current_state}' state.")
            return False

# --- Simulation of the Industrial Gap ---

firewall = IndustrialFirewall()

# 1. Manual Operation
normal_traffic = ["STARTING", "RUNNING", "OPEN_VALVE", "CLOSE_VALVE", "STOPPING", "OFF"]

# 2. Malicious AI/Bot Attack (Trying to open valve while system is OFF)
attack_traffic = ["OPEN_VALVE"]

print("--- Starting Normal Operations ---")
for cmd in normal_traffic:
    time.sleep(0.5)
    firewall.inspect_packet(cmd)

print("\n--- Starting Malicious/AI Injection Test ---")
time.sleep(1)
is_safe = firewall.inspect_packet("OPEN_VALVE")

if not is_safe:
    print("[FIREWALL] Packet Dropped. System Integrity Maintained.")
```
